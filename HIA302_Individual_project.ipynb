```python
#l: Comments and execution of the code in Figure 1

from numpy import isnan
from pandas import read_csv
from sklearn.impute import SimpleImputer

#Missing code in line 5 and 6: loading dataset and replacing '?' with NaN

dataframe = read_csv('C:\\Users\\HP\\OneDrive\\Desktop\\HIA302 IndProject\\HIA302_IndProject\\horse-colic.csv', header=None, na_values=['?'])
dataframe
```

```python
#Comment for line 8: Retrieving the values in the data frame 

data = dataframe.values
data
```

```python
#Comment for line 11: 

#: Retrieving the number of columns in  the data frame
totalCol = data.shape[1] 

#: Creating arrays of X and Y 
ix=[i for i in range(totalCol) if i !=23]
X = data[:, ix]
y = data[:, 23]
X, y
```

```python
#Comment for line 17: Printing the total missing values in the data

print('Missing: %d' % sum(isnan(X).flatten()))
```

```python
#Comment for line 20: Defining the imputer

imputer = SimpleImputer(strategy='mean')
```

```python
#Comment for line 22: Fitting the imputation on X array

imputer.fit(X)
Xtrans = imputer.transform(X)
```

```python
#Comment for line 26: Printing the total missing values after imputation 

print('Missing: %d' % sum(isnan(Xtrans).flatten()))
```
